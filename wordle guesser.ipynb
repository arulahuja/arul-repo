```python
five = [] 
```

```python
shared_dict = {}
```

```python
def cwc(word, word_2): #Fuction that uses Counter to discern the amount of common words
    common_letters = Counter(word) & Counter(word_2)
    return(sum(common_letters.values()))
```

```python
for word in five: #forloop to check for commonality
    sumofcommon = 0
    for word_2 in five:
        if word == word_2: #skip if word is the same 
            pass
        elif cwc(word,word_2)>0: #if has common, enter this
            sumofcommon = sumofcommon+cwc(word,word_2) #Sum the commons
    shared_dict[word] = sumofcommon #Assign commonality to the words
            
```

```python
def wordindexer(letter, index, chars):
    change = list(chars)
    change[index] = letter
    str1 = ''.join(change)
    return(str1)
        
```

```python
def guesser(word1, mystery_word,prev):
    listofcorrect = []
    listnot = []
    flag = False
    for letter1 in word1:
        for letter2 in mystery_word:
            if letter1 == letter2:
                if word1.index(letter1) == mystery_word.index(letter2):
                    guessed_word = wordindexer(letter1, word1.index(letter1),prev)
                    prev = guessed_word
                    flag = True
                if letter1 not in listofcorrect:
                        listofcorrect.append(letter1)
            elif letter1 not in list(mystery_word):
                if letter1 not in listnot:
                    listnot.append(letter1)
        if flag is False:
            guessed_word = prev
    return (guessed_word, listofcorrect, listnot)
    
```

```python
def listupdater(word_freq, results):
    updated_list = []
    indexletter = {}
    for letters in results[0]:
        if letters!='x':
            indexletter[results[0].index(letters)] = letters
    for k,v in word_freq:
        if set(results[1]).issubset(list(k)):
            #if set(results[2]).issubset(k) is False:
            if any(item in list(results[2]) for item in list(k)) is False:
                if results[0] != 'xxxxx':
                    check = 0
                    for index, l in indexletter.items():
                        if k[index] == l:
                            check = check+1
                    if check == len(indexletter):
                        updated_list.append((k,1))
                else:    
                    updated_list.append((k,1))
    return updated_list
```
