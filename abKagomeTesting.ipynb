```python
import networkx as nx
import scipy as sp
import numpy as np
```

```python
def buildKagome(n,a,b):
    G = nx.Graph()
    for i in range(n):
        i = i*5
        G.add_node(i+1)
        G.add_node(i+2)
        G.add_node(i+3)
        G.add_node(i+4)
        G.add_node(i+5)
        
        G.add_edge(i+1, i+2, weight=a)
        G.add_edge(i+1, i+5, weight=a)
        G.add_edge(i+2, i+5, weight=a)
        G.add_edge(i+3, i+4, weight=b)
        G.add_edge(i+3, i+5, weight=b)
        G.add_edge(i+4, i+5, weight=b)
        
        if(i>0):
            G.add_edge(i+1, i-3, weight=b)
            G.add_edge(i+3,i-1, weight=a)
            
    return G
    
```

```python
res = [1]
a = 1 #Vary a,b 
b = 1
print(((a*b)**2)*(a+b)) #Compare this to the actual growth rate
for n in range(1,100):
    G = buildKagome(n,a,b)
    #nx.draw(G)
    L = nx.laplacian_matrix(G).asfptype()
    Lr = sp.sparse.csr_matrix.todense(L[0:-1,0:-1])
    newRes = sp.linalg.det(Lr)
    #print(newRes)
    print((newRes/res[-1])/20.52)
    res = res + [newRes]
```
